```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm, datasets
import xgboost as xgb
%matplotlib inline
```

```python
train_fuel = pd.read_csv('train1_f.csv')
test_fuel = pd.read_csv('test1_f.csv')
```

```python
train_fuel
```

```python
train_fuel = train_fuel.loc[train_fuel['PH']==5]
```

```python
train_fuel
```

```python
test_fuel = test_fuel.loc[test_fuel['PH']==5]
```

```python
test_fuel
```

```python
feature_columns_to_use = ['ELEV_1',	'ELEV_2',	'FLAP',	'FQTY_1',	'FQTY_2',	'FQTY_3',	'FQTY_4',	'HDGS',	'OIT_1',	'OIT_2',	'OIT_3',	'OIT_4',	'PTRM',	'SAT',	'TAT',	'VMODE',	'VSPS',	'RUDD_Mean',	'RUDP_Mean',	'ALT_Mean',	'ALTR_Mean',	'AOA1_Mean',	'AOA2_Mean',	'AOAC_Mean',	'AOAI_Mean',	'CAS_Mean',	'GS_Mean',	'MACH_Mean',	'N1T_Mean',	'TAS_Mean',	'TRK_Mean',	'TRKM_Mean',	'VIB_1_Mean',	'ROLL_Mean',	'AIL_1',	'AIL_2',	'A_T', 'PH']
```

```python
big_X = train_fuel[feature_columns_to_use].append(test_fuel[feature_columns_to_use])
```

```python
big_X
```

```python
big_X = big_X.loc[big_X['PH']==5]
```

```python
big_X
```

```python
train_X = big_X[0:train_fuel.shape[0]].as_matrix()
test_X = big_X[train_fuel.shape[0]::].as_matrix()
train_y = train_fuel['FF']
```

```python
len(train_X)
```

```python
len(test_X)
```

```python
gbm = xgb.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=1, booster='gbtree').fit(train_X,train_y)
```

```python
predictions = gbm.predict(test_X)
```

```python
predictions
```

```python
test_y = test_fuel['FF']
```

```python
from sklearn.metrics import explained_variance_score
```

```python
print(explained_variance_score(predictions,test_y))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(predictions,test_y)
```

```python
print("Mean Absolute Error: ", np.sqrt(((predictions - test_y) ** 2).mean()))
```

```python
def mean_absolute_percentage_error(test_y, predictions): 
    test_y, predictions = np.array(test_y), np.array(predictions)
    return np.mean(np.abs((test_y - predictions) / test_y)) * 100
```

```python
print("Mean Absolute Percentage Error: ", np.mean(np.abs((test_y - predictions) / test_y)) * 100)
```
